```python
import pandas as pd
from TwitterUser import TwitterUser
```

```python
# Get the updated List from wikipedia
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_current_United_States_senators#G')[5]
df.head()
```

```python
#### 2- Search all using twitter API, it is very possible that some accounts are not found or wrongly found,
####but we are searching for nodes, not accuracy

# Define a twitter user instance
user = TwitterUser()

# Search each user on twitter and build a dataframe
user_dictionary = {}
for person in df.Senator.to_list():
    try:
        user_dictionary[person] = user.search_user(person)
    except:
        user_dictionary[person] = None
user_dictionary
```

```python
# An example
user_dictionary['Lisa Murkowski']
```

```python
# Now we get the keys that we want for our final dataset
preferred_keys = ['id', 'name', 'screen_name', 'location', 'followers_count', 'friends_count']

# And we extract them from the dictionary, we will make a simple function to protect from None values
def extract_key(user, key):
    try:
        return user_dictionary[user][key]
    except:
        return ''

for key in preferred_keys:
    df[key] = df.Senator.apply(lambda x: extract_key(x, key))

df.head(6)
```

```python
# See None values, to check if the dataset is acceptable
df[df.id == '']
```

```python
# Because every senator has a Twitter account we already have the network. Now a bit of cleaning
df.drop(['Portrait', 'Party'], axis=1, inplace=True)
df.rename({'Party.1': 'Party'}, axis=1, inplace=True)
df.head()
```

```python
df.rename({'Born': 'Age'}, axis=1, inplace=True)
df.Age = df.Age.apply(lambda x: int(x[5:7]))
```

```python
# This is surprising
df.Age.mean()
```

```python
# Save the dataset
df.to_csv('us_senators.csv', index = False)
```
